# Data Preprocessing & EDA

```python
tuna.describe()
```

```python
#Dropping the Domain column
dfsa = tuna.drop(['Domain'], axis = 1).copy()
```

```python
#checking the data for null or missing values
dfsa.isnull().sum()
```

```python
# shuffling the rows in the dataset so that when splitting the train and test set are equally distributed
dfsa = dfsa.sample(frac=1).reset_index(drop=True)
dfsa.head()
```

# Splitting the Data

```python
# Sepratating & assigning features and target columns to X & y
y = dfsa['Label']  #target variable
X = dfsa.drop('Label',axis=1)   #independent variable
```

```python
# Checking the shape for X, y variable 
X.shape, y.shape
```

```python
tuna['Label'].value_counts()
```

```python
# Splitting the dataset into train and test sets: 80-20 split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 12)  #test-size 20%
X_train.shape, X_test.shape
```

```python
# Creating holders to store the model performance results
ML_Model = []
acc_train = []
acc_test = []

#function to call for storing the results
def storeResults(model, a,b):
  ML_Model.append(model)
  acc_train.append(round(a, 3))
  acc_test.append(round(b, 3))
```

### Autoencoder Neurral Network

```python
#importing required packages
import keras
from keras.layers import Input, Dense
from keras import regularizers
import tensorflow as tf
from keras.models import Model
from sklearn import metrics
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#building autoencoder model

input_dim = X_train.shape[1]
encoding_dim = input_dim

input_layer = Input(shape=(input_dim,))
encoder = Dense(encoding_dim, activation='relu',
                activity_regularizer=regularizers.l1(10e-4))(input_layer)
encoder = Dense(int(encoding_dim - 2), activation='relu')(encoder)
code = Dense(int(encoding_dim - 4), activation='relu')(encoder)

# Decoder
decoder = Dense(int(encoding_dim - 2), activation='relu')(code)
decoder = Dense(encoding_dim, activation='relu')(decoder)
output_layer = Dense(input_dim, activation='sigmoid')(decoder)
autoencoder = Model(inputs=input_layer, outputs=output_layer)
autoencoder.summary()
```

```python
#compiling the model
autoencoder.compile(optimizer='adam',
                    loss='mean_squared_error',
                    metrics=['accuracy'])

# Set up EarlyStopping callback
early_stopping = EarlyStopping(
    monitor='val_accuracy',     # Monitor validation accuracy
    patience=3,                 # Number of epochs with no improvement after which training will be stopped
    mode='max',                 # We want to maximize the validation accuracy
    restore_best_weights=True   # Restore model weights from the epoch with the best value of the monitored quantity
)

#Training the model
history = autoencoder.fit(X_train, X_train, epochs=10, batch_size=64, shuffle=True, validation_split=0.2, callbacks=[early_stopping])  # Include the EarlyStopping callback 
```

```python
# Plotting the training and validation loss
# Create a figure with two subplots
fig, ax = plt.subplots(2, 1, figsize=(8, 6))

# Plot training and validation loss
ax[0].plot(history.history['loss'], label='Training Loss')
ax[0].plot(history.history['val_loss'], label='Validation Loss')
ax[0].set_title('Training and Validation Loss')
ax[0].set_xlabel('Epoch')
ax[0].set_ylabel('Loss')
ax[0].legend()

# Plot training and validation accuracy
ax[1].plot(history.history['accuracy'], label='Training Accuracy')
ax[1].plot(history.history['val_accuracy'], label='Validation Accuracy')
ax[1].set_title('Training and Validation Accuracy')
ax[1].set_xlabel('Epoch')
ax[1].set_ylabel('Accuracy')
ax[1].legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
acc_train_auto = autoencoder.evaluate(X_train, X_train)[1]
acc_test_auto = autoencoder.evaluate(X_test, X_test)[1]

print('\nAutoencoder: Accuracy on training Data: {:.3f}' .format(acc_train_auto))
print('Autoencoder: Accuracy on test Data: {:.3f}' .format(acc_test_auto))
```

```python
#storing the results. The below mentioned order of parameter passing is important.
storeResults('AutoEncoder', acc_train_auto, acc_test_auto)
```

### XGBoost Classifier

```python
#XGBoost Classification model
from xgboost import XGBClassifier

# instantiate the model
xgb = XGBClassifier(learning_rate=0.4,max_depth=7)
#fit the model
xgb.fit(X_train, y_train)
```

```python
#predicting the target value from the model for the samples
y_test_xgb = xgb.predict(X_test)
y_train_xgb = xgb.predict(X_train)
```

```python
#computing the accuracy of the model performance
from sklearn.metrics import accuracy_score
acc_train_xgb = accuracy_score(y_train,y_train_xgb)
acc_test_xgb = accuracy_score(y_test,y_test_xgb)

print("XGBoost: Accuracy on training Data: {:.3f}".format(acc_train_xgb))
print("XGBoost : Accuracy on test Data: {:.3f}".format(acc_test_xgb))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# Confusion Matrix
plt.figure(figsize=(8, 3))
plt.subplot(1, 2, 1)
sns.heatmap(confusion_matrix(y_train, y_train_xgb), annot=True, fmt = 'd',cmap='Greens')
plt.title("Training Confusion Matrix")
plt.xlabel("Predicted labels")
plt.ylabel("True labels")

plt.subplot(1, 2, 2)
sns.heatmap(confusion_matrix(y_test, y_test_xgb), annot=True, fmt = 'd',cmap='Greens')
plt.title("Test Confusion Matrix")
plt.xlabel("Predicted labels")
plt.ylabel("True labels")

plt.subplots_adjust(wspace=1)

plt.show()

# Classification Report
print("Training Classification Report:")
print(classification_report(y_train, y_train_xgb))
print("Test Classification Report:")
print(classification_report(y_test, y_test_xgb))
```

```python
#storing the results. The below mentioned order of parameter passing is important.
#Caution: Execute only once to avoid duplications.
storeResults('XGBoost', acc_train_xgb, acc_test_xgb)
```

### Multilayer Perceptron

```python
# Multilayer Perceptrons model
from sklearn.neural_network import MLPClassifier

# instantiate the model
mlp = MLPClassifier(alpha=0.001, hidden_layer_sizes=([100,100,100]))

# fit the model 
mlp.fit(X_train, y_train)
```

```python
#predicting the target value from the model for the samples
y_test_mlp = mlp.predict(X_test)
y_train_mlp = mlp.predict(X_train)
```

```python
#importing packages
from sklearn.metrics import accuracy_score
#computing the accuracy of the model performance
acc_train_mlp = accuracy_score(y_train,y_train_mlp)
acc_test_mlp = accuracy_score(y_test,y_test_mlp)

print("Multilayer Perceptrons: Accuracy on training Data: {:.3f}".format(acc_train_mlp))
print("Multilayer Perceptrons: Accuracy on test Data: {:.3f}".format(acc_test_mlp))
```

```python
# Visualizing the MLP architecture
plt.figure(figsize=(6, 3))
plt.title("Multilayer Perceptron Architecture")
plt.xlabel("Layers")
plt.ylabel("Neurons")
layers = [X_train.shape[1]] + list(mlp.hidden_layer_sizes) + [len(np.unique(y_train))]
plt.bar(range(len(layers)), layers)
plt.xticks(range(len(layers)), ["Input"] + ["Hidden {}".format(i+1) for i in range(len(mlp.hidden_layer_sizes))] + ["Output"])
plt.show()
```

```python
# Visualizing the accuracy scores
plt.figure(figsize=(6, 3))
plt.title("Multilayer Perceptron Accuracy")
plt.xlabel("Data")
plt.ylabel("Accuracy")
plt.bar(["Training", "Test"], [acc_train_mlp, acc_test_mlp], width=0.5)
plt.show()
```

```python
#storing the results. The below mentioned order of parameter passing is important.
#Caution: Execute only once to avoid duplications.
storeResults('Multilayer Perceptrons', acc_train_mlp, acc_test_mlp)
```
